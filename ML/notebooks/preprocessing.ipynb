## Encoding / Preprocessing	
Encode Categorical Variable (Sex Column) by converting it into numerical format as (female= 1, male= 0).
Missing values in Embarked were filled with the most common port.
The Embarked ports were encoded as numbers from 1 to 3 to make them usable for the model.


```python
import pandas as pd 
```

We first load the training and testing datasets.

```python
train_df = pd.read_csv("../../Data/test.csv")
test_df = pd.read_csv("../../Data/test.csv")

train_df.head()
test_df.head()
```

Encode the Sex Column: 
Requirement: Female = 1 and Male = 0

```python
sex_mapping = {'female': 1, 'male': 0}

train_df['Sex'] = train_df['Sex'].map(sex_mapping)
test_df['Sex'] = test_df['Sex'].map(sex_mapping)
```

```python
train_df.head()
#test_df.head()
```

## Train set:  Embarked column has (2) missing values.
Find the Most Common Embarked Port
Requirement: Missing Embarked values must be filled with the most common port.


```python
most_common_port = train_df['Embarked'].mode()[0]
print("Most common Embarked port:", most_common_port)
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna(most_common_port)
```

Encode Embarked as Numbers (1 to 3)
Requirement:
•	S (Southampton) = 1
•	C (Cherbourg) = 2
•	Q (Queenstown) = 3

```python
embarked_mapping = {
    'C': 1,
    'Q': 2,
    'S': 3
}

train_df['Embarked'] = train_df['Embarked'].map(embarked_mapping)
```
